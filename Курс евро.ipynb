```python
import requests
from datetime import date

def get_exchange_rates():
    # Вставьте свой ключ API здесь
    api_key = "0eddfb8689464c568d5ad81ed68e524e"

    # Получить сегодняшнюю дату
    today = date.today()

    # URL API
    url = f"https://openexchangerates.org/api/latest.json?app_id={api_key}"

    # Отправить запрос GET
    response = requests.get(url)

    # Преобразовать ответ в JSON
    data = response.json()

    # Проверить, есть ли в ответе ключ 'rates'
    if 'rates' in data:
        # Получить курсы валют
        rates = data['rates']

        # Вывести курсы валют
        for currency, rate in rates.items():
            print(f"1 USD = {rate} {currency}")
    else:
        # Вывести содержимое ответа для дальнейшего анализа
        print(data)

get_exchange_rates()
```
